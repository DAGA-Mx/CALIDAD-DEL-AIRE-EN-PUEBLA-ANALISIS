#### QUE SE LIMPIARAN Y ANALIZARÁN HACIENDO USO DE LA LIBRERÍA 'PANDAS'

## Una vez identificados procedemos a mitigarlos con valores del promedio procedientes de la misma columna

```python
# lOS DATOS "nan" son reemplazados por el dato anterior
df['Estacion'] = df['Estacion'].fillna(method="ffill")
df
```

# SE REEMPLAZARÁN LOS DATOS VACÍOS POR EL VALOR DE FILA ANTERIOR PARA LAS COLUMNAS STRING/CATEGÓRICAS:

```python
# lOS DATOS "nan" son reemplazados por el dato anterior
df['Agrupacion'] = df['Agrupacion'].fillna(method="ffill")

df['Fecha inicial'] = df['Fecha inicial'].fillna(method="ffill")

df['Fecha final'] = df['Fecha final'].fillna(method="ffill")

df['Parametro'] = df['Parametro'].fillna(method="ffill")
```

```python
# Función para reemplazar "Auto%#" por el valor anterior en cada columna
def reemplazar_auto(df):
    return df.apply(lambda col: col.mask(col == "Auto%#").ffill())

# Aplicar la función
df_limpio = reemplazar_auto(df)

df_limpio
```

# CREACIÓN DE UNA LISTA PARA LAS COLUMNAS NUMÉRICAS:
## REEMPLAZO DE LOS DATOS VACÍOS Y TEXTO ATÍPICO CON VALOR PROMEDIO DE LA COLUMNA

```python
print(df.columns.tolist())
```

```python
# Lista de columnas numéricas
columnas_numericas = [
    'Dias buenos',
    'Dias aceptables',
    'Dias malos',
    'Dias muy malos',
    'Dias extremadamente malos',
    'Dias insuficientes'
]

# Reemplazar "Auto%#" por NaN para poder calcular el promedio
df[columnas_numericas] = df[columnas_numericas].replace("Auto%#", pd.NA)

# Convertir las columnas a tipo numérico (por si tienen texto)
df[columnas_numericas] = df[columnas_numericas].apply(pd.to_numeric, errors='coerce')

# Rellenar los NaN con el promedio de cada columna
df[columnas_numericas] = df[columnas_numericas].apply(lambda col: col.fillna(col.mean()))
```
